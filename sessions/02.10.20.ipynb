```python
A = np.array([[0, 0], [0, 1], [2, 0]])
```

```python
A
```

R(alpha) = \[\[cos(alpha), -sin(alpha)], [sin(alpha), cos(alpha)\]\]

A * R(alpha) rotates A by alpha degrees

```python
import math as m
```

```python
R45 = np.array([[m.sqrt(2)/2, -m.sqrt(2)/2], [m.sqrt(2)/2, m.sqrt(2)/2]])
```

```python
A @ R45
```
